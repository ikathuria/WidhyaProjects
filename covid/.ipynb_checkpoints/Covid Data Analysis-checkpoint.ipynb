```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
covid_data = pd.read_csv('covid19.csv', index_col=0)
covid_data.head()
```

```python
new_covid = covid_data.groupby(['Date'], sort=False).sum()
new_covid
```

```python
total_cases = new_covid.sum(axis=1)
total_cases
```

```python
fig, ax = plt.subplots(figsize=(50, 6))
ax.plot(new_covid.index,total_cases);
```

```python
dates = total_cases['04/03/20':'21/03/20']
dates
```

```python
values_of_r = []

for i in range(len(dates)-1):
    values_of_r.append((dates[i+1] - dates[i]) / dates[i])

values_of_r
```

```python
r = sum(values_of_r) / len(values_of_r)
```

```python
P_o = 31
t = 26
```

```python
import math
P_t = P_o * (math.exp(r*t))
```

```python
round(P_t)
```
